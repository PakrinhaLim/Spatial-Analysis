```python
def readb64(base64_string):
    nparr = np.frombuffer(base64.b64decode(base64_string), np.uint8)
    try:
        img = cv2.imdecode(nparr, cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    except:
        img = cv2.imdecode(nparr, cv2.COLOR_RGB2BGR)
        return img
    return gray
```

```python
def imgToArr(image, dimension):
    pixelData = []
    for l, data in enumerate(image):
        pixelData.append([])
        for i in range(dimension[0]):#dimension[0]
            for j in range(dimension[1]):#dimension[1]
                pixelData[l].append(data[i, j])
    return pixelData
```

```python
def findMaxMin(image):
    data = []
    max = []
    min = []
    for item in image:
        data.append(imgToArr(item, item[0].shape))
    for i in range(len(data[0])):
        new_data = [item[i] for item in data]
        max.append(np.amax(new_data))
        min.append(np.amin(new_data))
    return [max]+[min]
```

```python
def imageTransform(image):
    data = []
    for item in image:
        data.append(imgToArr(item, item[0].shape))
        data[-1] = np.array(data[-1]).transpose()
    return data
```

```python
def normalize(dna, max, min):
    new_dna = []
    avg = np.subtract(max,min)
    for i, data in enumerate(dna):
        for item in data:
            a = np.true_divide((item-min),avg)
            b = np.array([int(i+1)])
            new_dna.append(np.insert(a, len(item) , b))
    return new_dna
```

```python
def initPop(dna, npop):
    new_pop = []
    for i in range(npop):
        data = []
        for j in range(3):
            num1 = (len(dna)//3)*j
            num2 = (len(dna)//3)*(j+1)
            # random data from A B C image
            data.append([
                        random.choice([data[0] for data in dna[num1:num2]]), # 1st band
                        random.choice([data[1] for data in dna[num1:num2]]), # 2nd band
                        random.choice([data[2] for data in dna[num1:num2]]),
                        random.choice([data[3] for data in dna[num1:num2]]),
                        random.choice([data[4] for data in dna[num1:num2]]),
                        random.choice([data[5] for data in dna[num1:num2]]),
                        random.choice([data[6] for data in dna[num1:num2]]), # 7th band
                        random.choice([data[7] for data in dna[num1:num2]])
                        ])
        new_pop.append(data)

    return new_pop
        
```

```python
def createTestAndTrainData(image, npop, max, min):
    testData = []
    trainData = []
    data = imageTransform(image)
    for i in range(npop):
        test = []
        train = []
        for item in data:
            new_data = np.array(item)
            max, min = np.array(max), np.array(min)
            new_data = (new_data-min)/(max-min)
            num1 = (len(new_data)//npop) * i
            num2 = (len(new_data)//npop) * (i+1)
            test.append(new_data[num1:num2])
            train.append(np.concatenate((new_data[0:num1],new_data[num2:len(new_data)])))
        testData.append(test)
        trainData.append(train)

    return [testData]+[trainData]
```

```python
def evaluateObj(dna, trainData, npop):
    count = []
    for item in dna: #10
        cnt = 0
        for i, subitem in enumerate(item):#3 By class
            for j in range(len(trainData[0])):#3626 #3->3626->7
                Dp = []
                for k in range(3):
                    distance = np.sqrt(np.sum(np.power(np.subtract(subitem[:-1],trainData[k][j]),2)))
                    Dp.append(np.sum(distance))
                minIndex = Dp.index(np.amin(Dp))
                if(subitem[-1] == (minIndex+1)):
                    cnt += 1
        count.append(cnt)
    return count
```

```python
def roulette(dna, evaluateValue, npop):
    calculate = 0
    weight = []
    for item in evaluateValue:
        calculate += np.true_divide(item, np.sum(evaluateValue))
        weight.append(calculate)
    if(weight[-1] != 1):
        weight[-1] = 1
    data = []
    for i in range(npop):
        k = random.random()
        for j, item in enumerate(weight):
            if k <= item:
                data.append(dna[j].copy())
                data[-1].append([item])
                break
    return data
```

```python
def crossOver(dna, npop, gene):
    Pc = 0.8
    new_dna = []
    dnaIndex = [data for data in range(npop)]
    for i in range(len(dna)//2):
        matchmaker = random.sample(dnaIndex, 2)
        child_1 = [data for data in dna[matchmaker[0]][:-1]]
        child_2 = [data for data in dna[matchmaker[1]][:-1]]
        dnaIndex.remove(matchmaker[0])
        dnaIndex.remove(matchmaker[1])
        data1 = []
        data2 = []
        if random.random() < Pc:
            cut = random.choice([data for data in range(1, gene-1)])
            for j in range(3):
                data1.append([data for data in child_1[j][:int(cut)]] + [data for data in child_2[j][int(cut):]])
                data2.append([data for data in child_2[j][:int(cut)]] + [data for data in child_1[j][int(cut):]])
        else:
            data1.append(child_1)
            data2.append(child_2)
            data1 = [subdata for data in data1 for subdata in data]
            data2 = [subdata for data in data2 for subdata in data]
        new_dna.append(data1)
        new_dna.append(data2)
    return new_dna
```

```python
def mutation(dna, npop, gene):
    Pm = 0.2
    for item in dna:
        if(random.random() < Pm):
            for subitem in item:
                k = random.uniform(-0.05,0.05)
                for i in range(len(subitem)-1):
                    subitem[i] += float(k)
                    if subitem[i] > 1 :
                        subitem[i] = 1
                    if subitem[i] < 0 :
                        subitem[i] = 0
    return dna
```

```python
def selection(child_dna, child_evaluateValue, parent_dna, parent_evaluateValue, npop):
    dna = child_dna + parent_dna
    evaluateValue = child_evaluateValue + parent_evaluateValue
    orderIndex = np.argsort(-np.array(evaluateValue), kind='mergesort')
    evaluateValue.sort(reverse = True,key=lambda x: x)
    new_dna = [dna[i] for i in orderIndex]
    return new_dna[:npop], evaluateValue[:npop]
```
